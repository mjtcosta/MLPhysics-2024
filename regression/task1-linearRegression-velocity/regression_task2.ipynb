# Exemplo de gŕafico

```python
# Dados para exemplo de gráfico e  regressão linear.
X = np.array([[1], [2], [3], [4], [5]])  # Input features 
y = np.array([11, 23, 29, 41, 52])       # Target

# Exemplo de Gráfico

plt.figure(figsize=(8, 5))

plt.plot(X,y, 'o-', color='blue',label='Real')        # "o-" Significa gráfico com pontos e ligados por uma linha.


plt.title('Titulo de Gráfico')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.grid(True)
plt.legend()
#plt.savefig('nome_gráfico.png', dpi=300, bbox_inches='tight', transparent=True) # Salvando gráfico para um arquivo.
plt.show()

```

# Exemplo de Regressão Linear usando o scikit-learn.

```python


model = LinearRegression()               # Inclua um cometário sobre o algoritimo de regressão linear.

model.fit(X,y)                           # Obtendo os parâmetros via Regressão Linear. 

# y = A + B*X
B = model.coef_                          # Inclua um comentário sobre a interpretação do parâmetro B
A = model.intercept_                     # Inclua um comentário sobre a interpretação do parâmetro A

print("Coeficiente:", A)
print("Intercepto:", B)


y_pred = np.zeros_like(y)

# Valored preditos pelo modelo.
y_pred[:] = A + B*X[:,0]


 
```

# Exemplo de Gráfico comparando os valores reais e preditos.

```python
# Novo Gráfico com os pontos Reais e os Preditos pela regressão.
plt.figure(figsize=(8, 5))

plt.plot(X,y, 'o', color='blue',label='Real')        # Gráfico dos valores reais
plt.plot(X,y_pred, '-', color='red',label='Predito') # Gráfico dos valores preditos pelo modelo



plt.title('Titulo de Gráfico')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.grid(True)
plt.legend()
#lt.savefig('nome_gráfico.png', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```
